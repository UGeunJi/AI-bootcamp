# 신경망 

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

## 활성화 함수

* 계단함수

```python
def step_function(x):
    return np.array(x > 0, dtype=np.int32)
```

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = step_function(X)

plt.plot(X, Y)
```

* 시그모이드 함수

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = sigmoid(X)

plt.plot(X, Y)
```

* ReLU

```python
def relu(x):
    return np.maximum(0, x)
```

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = relu(X)

plt.plot(X, Y)
```
