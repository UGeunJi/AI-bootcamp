# 정규식(Regular Expression)
```
 "특정 조건 또는 패턴"을 치환하는 과정을 쉽게 처리할 수 있는 방법
```

- match() : 문자열의 첫 시작부터 정규식과 매치되는지 조사한다.
- search() :	문자열 전체를 검색하여 정규식과 매치되는지 조사한다.
- findall()	: 정규식과 매치되는 모든 문자열(substring)을 리스트로 돌려준다.
- finditer() : 정규식과 매치되는 모든 문자열(substring)을 반복 가능한 객체로 돌려준다.
https://docs.python.org/ko/3/library/re.html?highlight=re#module-re

re 패키지를 사용하는 법은 2가지가 있다
```
p = re.compile('패턴')
p.method(데이터)
```

### [문자 클래스]

1. [abc] - a, b, c중 한 개의 문자와 매치
  - a : 매치여부 OK
  - apple : ok
  - double : no

2. \d   : 숫자와 매치, [0-9]와 동일

3. \D : 숫자가 아닌 것과 매치 [^0-9]와 동일

4. \s : whitespace 문자와 매치

5. \S : whitespace 문자가 아닌것과 매치, 

6. \w : 문자 + 숫자와 매치, [a-zA-Z0-9]와 동일

7. \W : 문자+숫자가 아닌 문자와 매치. [^a-zA-Z0-9]와 동일
  


8. a.b : a와 b 사이에 줄바꿈 문자를 제외한 모든 문자 허용

9. a[.]b : a와 b 사이에 dot 문자만 허용

10. ca*t  : a 문자가 0번 이상 반복 허용

11. ca+t :  a 문자가 1번 이상 반복 허용

12. ca?t : a 문자가 없거나, 1번만 허용

13. ca{3}t : a 문자가 3번 반복되면 매치

14. ca{2, 3}t : a 문자가 2~3번 반복되면 매치

````python
str4 = "string cast cat capital caaaat ct cat"
````

````python
re.findall('ca*t', str4)   # c로 시작하고 t로 끝나면서 a가 0번 이상 반복되는 모든 단어 출력
````

````python
#ca로 시작하고 t로 끝나는 모든 단어를 찾아보세요 
re.findall('ca[a-zA-Z]*t', str4)  
````

### 그루핑 패턴 
- 자주 사용하는 패턴은 별명을 지어 부를 수도 있습니다. 

````python
p = re.compile(r"(?P<name>\w+)\s+(?P<num>\d+[-]\d+[-]\d+)") # ?P<name>\w+ -  \w+ 를 name이라는 패턴명으로 부르기로 했습니다
# (?P<name>\w+) - name이라는 그룹명으로 \w+
# (?P<num>\d+[-]\d+[-]\d+) - num이라는 그룹명으로 숫자-숫자-숫자 
print(p.sub("\g<num> \g<name>", "kim 010-1234-1234")) # \g<phone> : phone이라는 그룹을 출력해주세요
````

````python
print(p.sub("\g<num>", "kim 010-1234-1234"))
````

````python
p = re.compile("(?P<name>\w+)\s+(?P<num>\d+[-]\d+[-]\d+)")
print(p.sub("\g<2> \g<1>", "kim 010-1234-1234"))  # 이름은 모르겠고 그룹중에 2번, 그룹중에 1번을 출력해주세요
````

````python
# 찾고(앞의 4가지 검색용 함수) -> 찾은 애를 변경할 때 - sub()
string = input('전화번호를 입력하세요')
p = re.compile('(?P<first_num>[0-9]{6}[-][0-9]{1})[0-9]{6}') # 패턴을 만들어준다
p.sub("\g<first_num>******", string)
````

````python
# 찾고(앞의 4가지 검색용 함수) -> 찾은 애를 변경할 때 - sub()
string = input('전화번호를 입력하세요')
p = re.compile('(?P<first_num>[0-9]{6}[-][0-9]{1})[0-9]{6}') # 패턴을 만들어준다
p.sub("\g<1>******", string)
````

````python
# 2.
# 아이디 부분은 : id 라는 그룹명
# 도메인주소: first 라는 그룹명
# .뒤의 마지막부분은 last라는 그룹명을 지어서 변경해보시고 
# id만 따로 출력을 해 보세요 
# pat = re.compile('\S+@\S+[.]\w*')  

pat = re.compile('(?P<id>\S+)[@](?P<first>\S+)[.](?P<last>\w*)')  
# 이메일리스트 전체 탐색 (반복)
for email in emails:
    # 형식에 일치하면 출력
    if bool(pat.match(email)):
        print(pat.sub('\g<id>' ,email))
````
